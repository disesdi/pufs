```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
# read in CRP data

df = pd.read_csv('CRPdataset.csv')

df
```

```python
X = df.drop('Label', axis=1)

y = df['Label']
```

```python
y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=29)
```

```python
from sklearn.metrics import accuracy_score
```

## Cohort 1

### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()

y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print('accuracy: %.2f%%' % (accuracy * 100.0))
```

### Support Vector Machine

```python
from sklearn import svm

svc_clf = svm.SVC()

svc_y_pred = svc_clf.fit(X_train, y_train).predict(X_test)

accuracy = accuracy_score(y_test, svc_y_pred)

print('accuracy: %.2f%%' % (accuracy * 100.0))
```

### XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()

xgb.fit(X_train, y_train).predict(X_test)

xgb_pred = xgb.predict(X_test)

accuracy = accuracy_score(y_test, xgb_pred)

print('accuracy: %.2f%%' % (accuracy * 100.0))
```

### Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_1 = RandomForestClassifier()

rfc_1 = rfc_1.fit(X_train, y_train)

rfc_1_pred = rfc_1.predict(X_test)

accuracy = accuracy_score(y_test, rfc_1_pred)

print('accuracy: %.2f%%' % (accuracy * 100.0))
```

## Cohort 2

### Random Forests 2

```python
rfc = RandomForestClassifier(n_estimators=99)

rfc = rfc.fit(X_train, y_train)

rfc_pred = rfc.predict(X_test)

accuracy = accuracy_score(y_test, rfc_pred)

print('accuracy: %.2f%%' % (accuracy * 100.0))
```

### XGBoost 2

```python
xgb_2 = XGBClassifier(max_depth=11)

xgb_2 = xgb_2.fit(X_train, y_train)

xgb_2_pred = xgb_2.predict(X_test)

accuracy = accuracy_score(y_test, xgb_2_pred)

print('accuracy: %.2f%%' % (accuracy * 100.0))
```
